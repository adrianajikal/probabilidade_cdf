# Probabilidade - Tarefa

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

```python
# Parametros
media = 15
desvio_padrao = 0.1

# Probabilidade acumulada com funcao CDF (Cumulative Distribution Function).
prob_acum = stats.norm.cdf(15.2, loc=media, scale=desvio_padrao)
prob_acima = 1 - prob_acum
print(f'Probabilidade de pesar mais que 15,2 kg: {prob_acima:.4f}')
```

2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# funo ppf (Percent Point Function)
percentil_95 = stats.norm.ppf(0.95, loc=media, scale=desvio_padrao)
print(f'Percentil 95% do peso: {percentil_95:.4f} kg')
```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

```python
Y=np.linspace(-6, 6, 200)
# Valores a serem analisados
valores = np.array([-2, -1, 0, 1, 2])

# Calcula CDF para cada valor de Y
cdf_normal = stats.norm.cdf(Y, loc=0, scale=1)
cdf_tstudent = stats.t.cdf(Y, df=5)
```

4. Faa uma visualizao grfica dos dados do item anterior.

```python
# Plotando as curvas
plt.figure
plt.plot(Y, cdf_normal, label='Normal Padro')
plt.plot(Y, cdf_tstudent, label='t-Student (df=5)', linestyle='--')

plt.xlabel('Valores de Y')
plt.ylabel('Probabilidade Acumulada (CDF)')
plt.title('Comparao entre Normal Padro e t-Student (df=5)')
plt.legend()
plt.grid(True)
plt.show()
```

5. 3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
Y=np.linspace(0, 1, 20)
# Seu cdigo comea aqui
mu = 0
variance = 1
df = 5 #t-studdent

# Percentil para Normal Padro (inversa da CDF) 
percentis_normal = stats.norm.ppf(Y, loc=0, scale=1)

# Percentil para t-Student com 5 graus de liberdade
percentis_tstudent = stats.t.ppf(Y, df=5)

print("Percentil - Normal Padro vs t-Student (df=5)\n")
for p, pn, pt in zip(Y, percentis_normal, percentis_tstudent):
    print(f"Y = {p:.2f}: Normal = {pn:.4f}, t-Student = {pt:.4f}")
```
